신한카드의 신용카드, 체크카드 이름 목록 웹 스크래핑 코드

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time
import csv
```

```python
service = Service(executable_path='C:/University/chromedriver.exe')

driver = webdriver.Chrome(service=service)
```

체크카드

```python
url = 'https://www.shinhancard.com/pconts/html/card/check/MOBFM282R11.html?crustMenuId=ms52'
driver.get(url)
time.sleep(5)

html=driver.page_source

soup = BeautifulSoup(html, 'html.parser')

# 카드 이름 요소 찾기
# card_name_elements = soup.find_all('a', attrs={'class': 'card_name'})

card_elements=soup.find_all('li')
result = []

for element in card_elements:
    img_tag = element.find('img')
    if img_tag:
        img_src = img_tag.get('src')
    a_tag = element.find('a', class_='card_name')
    if a_tag:
        card_name=a_tag.get_text(strip=True)
        href=a_tag.get('href')
        
        result.append((img_src,card_name,href))
        
for img_src, card_name, href in result:
    print(f"Image Source: {img_src}")
    print(f"Card Name: {card_name}")
    print(f"Href: {href}")
    print('-' * 40)

# for element in card_name_elements:
#     print(element.get_text())
#     print(element.get('href'))


# csv_file_path = 'CheckCard_names2.csv'
# with open(csv_file_path, mode='w', newline='',encoding='utf-8') as file:
#     writer = csv.writer(file)
#     #writer.writerow(['CardName','CardIMG','CardDescURL'])
    
#     for img_src, card_name, href in result:
#         print(f"Image Source: {img_src}")
#         print(f"Card Name: {card_name}")
#         print(f"Href: {href}")
#         print('-' * 40)
#         writer.writerow([card_name, img_src, href,"check"])
i = 281
csv_file_path = 'CheckCard_benefits.csv'
with open(csv_file_path, mode='w', newline='',encoding='utf-8') as file:
    writer = csv.writer(file,delimiter=';')
    for img_src, card_name, href in result:
    #카드 상세페이지는 https://www.shinhancard.com + 카드위치URL

    #아래는 신한 SOL트래블 체크의 예시
        url='https://www.shinhancard.com'+ href
        driver.get(url)
        time.sleep(5)
        html=driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        
    #혜택 요소 찾기
        benefits = soup.find_all('li', class_='icon')
        for benefit in benefits:
            print(benefit.get_text())
            writer.writerow([i,card_name,benefit.get_text()])
            i=i+1
            
    #benefit.get_text()를 메인 화면 - 카드 선택 시 하단에 띄워주면 된다
```

CSV 파일 데이터에서 큰따옴표 없애기

```python
import pandas as pd

# CSV 파일 읽기
df = pd.read_csv('CheckCard_benefits2.csv')

# 모든 문자열 열에서 " 기호를 제거
df = df.apply(lambda x: x.str.replace('"', '', regex=False) if x.dtype == "object" else x)

# 결과 확인
print(df.head())

# 필요한 경우 수정된 데이터프레임을 다시 CSV로 저장
df.to_csv('CheckCard_benefits3.csv', index=False)
```
