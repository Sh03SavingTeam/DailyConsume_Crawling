```python

service = Service(executable_path='E:/부트 캠프/DailyConsume_SC/chromedriver-win64/chromedriver-win64/chromedriver.exe')

driver = webdriver.Chrome(service=service)
```

```python
# MySQL 서버와 연결 -> !! 깃헙 push 전 암호화 필수 !!
mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="1234",
  database="dailyconsume",
  port="3307"
)

mycursor = mydb.cursor()

convenient_list=["cu", "gs25", "seven", "ministop", "emart24"]

for convenient in convenient_list:
    url = f'https://pyony.com/brands/{convenient}/?page=1'
    driver.get(url)
    time.sleep(5)
    html=driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    product_elements=soup.find_all("div", attrs={"class":"col-md-6"})

    for element in product_elements:
        img_tag=element.find("img")
        if convenient=='cu':
            prod_img="https:"+img_tag.get("src")
        else:
            prod_img=img_tag.get("src")
    
        product_name_tag=element.find("strong")
        product_name = product_name_tag.get_text()
    
        product_price_tag = element.find("i", class_="fa fa-coins text-warning pr-1")
        amount = product_price_tag.next_sibling.strip()
    
        product_category_tag = element.find("small", class_="float-right font-weight-bold")
        product_category = product_category_tag.get_text()
    
    
        product_content_tag = element.find("span", class_=f"badge bg-{convenient} text-white")
        product_content = product_content_tag.get_text()
        
        store_name_tag = element.find("small", class_=f"float-left font-{convenient}-logo font-weight-bold")
        store_name=store_name_tag = store_name_tag.get_text()
        print(store_name, product_name, amount, product_content, product_category, prod_img)

        sql = "INSERT INTO t_discounting_info (store_name, product_name, amount, product_content, product_category, prod_img) VALUES (%s, %s, %s, %s, %s, %s)"
        val = (store_name, product_name, amount, product_content, product_category, prod_img)
        mycursor.execute(sql, val)

    mydb.commit()
```
